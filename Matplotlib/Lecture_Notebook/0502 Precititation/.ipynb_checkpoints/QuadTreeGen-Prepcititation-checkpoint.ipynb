```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from quadtree_prep import Point, Rect, QuadTree
```

```python
from config import GLOBAL_GPSE_MIN, GLOBAL_GPSE_MAX, GLOBAL_GPSN_MIN, GLOBAL_GPSN_MAX
domain = Rect(GLOBAL_GPSE_MIN, GLOBAL_GPSN_MIN, GLOBAL_GPSE_MAX, GLOBAL_GPSN_MAX)
GLOBAL_GPSE_MIN, GLOBAL_GPSE_MAX, GLOBAL_GPSN_MIN, GLOBAL_GPSN_MAX
```

```python
PREP_MIN, PREP_MAX = 0, 100
```

```python
import random
random.seed(5)

N = 50

prep = [random.random()*PREP_MAX for i in range(N)]
lat = [random.random()*(GLOBAL_GPSE_MAX-GLOBAL_GPSE_MIN)+GLOBAL_GPSE_MIN for i in range(N)]
long = [random.random()*(GLOBAL_GPSN_MAX-GLOBAL_GPSN_MIN)+GLOBAL_GPSN_MIN for i in range(N)]
```

```python
coords = list(zip(lat, long, prep))
points = [Point(*coord) for coord in coords]
```

```python
points[0]
```

```python
len(points)
```

```python
qtree = QuadTree(domain, 3)
```

```python
for point in points:
    result = qtree.insert(point)

print('Number of points in the domain =', len(qtree))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
def draw_quadtree_prep():
    DPI = 600
    fig = plt.figure(figsize=(980/DPI, 700/DPI), dpi=DPI, constrained_layout=True)
    ax = fig.add_axes([0.14,0.18,0.70,0.70])

    ax.scatter([p.x for p in points], [p.y for p in points], s=2, zorder=5
               , facecolors='b', edgecolor='none', marker='*')
    
    qtree.draw(ax)

    ax.set_xlim(GLOBAL_GPSE_MIN, GLOBAL_GPSE_MAX)
    ax.set_ylim(GLOBAL_GPSN_MIN, GLOBAL_GPSN_MAX)

    ax.xaxis.set_major_locator(plt.FixedLocator(np.linspace(GLOBAL_GPSE_MIN, GLOBAL_GPSE_MAX, 30)))
    ax.yaxis.set_major_locator(plt.FixedLocator(np.linspace(GLOBAL_GPSN_MIN, GLOBAL_GPSN_MAX, 15)))
    ax.xaxis.set_major_formatter(plt.FormatStrFormatter('%0.3f'))
    ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%0.3f'))
    ax.tick_params(axis='x', labelsize=2.5, width=0.4, length=2,labelrotation=90., pad=2)
    ax.tick_params(axis='y', labelsize=2.5, width=0.4, length=2, pad=2)


    # add colorbar after creating ScalarMappable object artificially
    N = 21
    cmap = plt.get_cmap('cool')
    norm = matplotlib.colors.Normalize(vmin=PREP_MIN,vmax=PREP_MAX)
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    #sm.set_array([])

    cbar_ax = fig.add_axes([0.86, 0.18, 0.03, 0.65])
    cbar = fig.colorbar(sm, cax=cbar_ax
                        , ticks=np.linspace(PREP_MIN,PREP_MAX,N)
                        , boundaries=np.arange(-2.5, 105., 5))
    cbar.ax.tick_params(labelsize=2.0, width=0.4, length=1.3, direction='out', pad=2)

    plt.tight_layout()
    plt.savefig('prep-quad.png', DPI=600, bbox_inchies='tight', pad_inchies=0)
```

```python
draw_quadtree_prep()
```
