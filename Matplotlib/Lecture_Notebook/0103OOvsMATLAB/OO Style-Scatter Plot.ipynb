```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

### Scatter Plot with legends

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    ax1.plot(rng.rand(5), rng.rand(5), marker, \
             label="marker='{0}'".format(marker))

ax1.legend(numpoints=1)
ax1.set_xlim(0, 1.8)
```

### Scatter Plot with Colorbar

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100) #randn : standard normal
colors = rng.rand(100) #rand : random samples from a uniform dist [0,1)
sizes = 1000 * rng.rand(100)

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# how to generator colorbar data ?
mappable = ax1.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
fig.colorbar(mappable, ax=ax1)
```
