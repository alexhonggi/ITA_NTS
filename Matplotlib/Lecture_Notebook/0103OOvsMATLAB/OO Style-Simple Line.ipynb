```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
```

### Basic Line Plot

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.plot(x, np.sin(x))
ax1.plot(x, np.cos(x))
```

### Adjusting the Plot: Line Colors and Styles

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, x + 0, linestyle='solid')
ax1.plot(x, x + 1, linestyle='dashed')
ax1.plot(x, x + 2, linestyle='dashdot')
ax1.plot(x, x + 3, linestyle='dotted')

ax1.plot(x, x + 4, linestyle='-') # solid
ax1.plot(x, x + 5, linestyle='--') # dashed
ax1.plot(x, x + 6, linestyle='-.') # dashdot
ax1.plot(x, x + 7, linestyle=':') # dotted
```

### Adjusting the Plot:Axes Limits

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(x, np.sin(x))
ax1.set_xlim(-1, 11)
ax1.set_ylim(-1.5, 1.5)
```

### Labeling Plots

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x, np.sin(x))
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.set_title('A Sine Curve')
```
