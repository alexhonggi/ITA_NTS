# Multiple Subplot 를 다양한 방법으로 그려보자

### multiple subplots 를 그리는 4가지 방법
##### 1. plt.subplots()
##### 2. fig.add_subplot()
##### 3. fig.add_axes()
##### 4. fig.add_subplot() + GridSpec

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 0. Figure 객체와 Axes 객체
- Figure : 그래프들이 그려지는 캔버스, 여러 Subplot 들을 포함한다.
- Axes : 단일 그래프 (Subplot)

```python
fig = plt.figure(figsize=(6,6)) # figure 객체를 (6,6) 사이즈로 생성
ax = fig.add_subplot() # Axes 객체 (subplot) 하나를 생성
```

#### 실습1. type(fig), type(ax) 를 각각 실행하여 fig, ax 변수의 객체가 무엇인지 확인해 보자.

### 1. plt.subplots(nrows, ncols)
- Figure 객체와 Axes 객체를 동시에 생성가능
- nrows, ncold 인자를 통해서 생성할 Axes 배열의 개수를 지정가능 (미지정시 1개)

```python
fig, axes = plt.subplots() #Axes 그래프 1개를 생성
type(axes)
```

```python
fig, axes = plt.subplots(2,3) #Axes 배열 (2,3) 을 생성
type(axes)
```

#### 실습1. plt.subplots() 함수를 사용하여 (1,2) 배열의 Axes 를 생성하라.
- 행수 : 1
- 열수 : 2

### 2. fig.add_subplot(nrows,ncols,index)
- 먼저 Figure 객체를 생성 후 (fig)
- 한번에 하나의 Axes 객체 (subplot) 추가

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) # fig.add_subplot(1,1,1) 과 동일한 결과
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,2,1) # 1행, 2열짜리 배열을 만들고 그 중에 첫번째에 subplot 추가
```

```python
ax = fig.add_subplot(1,2,2) # 나머지 2열에 추가
fig # fig 를 추가해야 화면에 그래프가 보임 (figure 객체가 같은 cell 안에 있지 않기 때문)
```

#### 실습2. fig.add_subplot() 함수를 사용하여 2열 2행의 Axes 객체를 생성하라.

### 3. fig.add_axes([left, bottom, width, height])
- 직사각형 (rectangle)을 사용하여 절대좌표에 subplot을 직접 그리기 가능
- 좌표는 왼쪽 아래 (left, bottom) 이 (0.,0.) 이고 오른쪽 위 (right, top) 이 (1., 1.) 임

```python
fig = plt.figure()
ax = fig.add_axes([0.,0.,1.,1.])
```

```python
# 새로운 subplot 을 추가
ax2 = fig.add_axes([0.2,0.2,0.3,0.6])
fig
```

#### 실습3. fig.add_axes() 함수를 사용하여 (0.6, 0.5)를 (left, bottom)으로하고 (0.3, 0.15) 를 (가로너비,세로너비)로 하는 ax3를 위 그래프에 추가하라.

## 4. fig.add_subplot() + GridSpec
- 가상의 격자(Grid)를 설정하고 Grid 범위에 맞춰서 Axes 객체를 하나씩 추가

```python
import matplotlib.gridspec as gridspec

fig = plt.figure()
spec = gridspec.GridSpec(4, 1, wspace=0.4, hspace=0.3)

ax1 = fig.add_subplot(spec[:2, 0])
ax2 = fig.add_subplot(spec[-1, 0])
```

#### 실습4. GridSpec 은 3행 3열로 설정하고, ax1 은 [:2, :2] 에 위치시키고, ax2 는 [2,1], ax3 은 [2,2] 에 생성하라
