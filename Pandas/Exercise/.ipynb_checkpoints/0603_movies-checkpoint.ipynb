## 1. 데이터 분석을 위한 데이터 정리

```python
df.columns
```

```python
# 미사용 컬럼 삭제
# Poster_Link, No_of_Votes, Gross, Meta_score, Overview
df_d = df.drop(['Poster_Link', 'No_of_Votes', 'Gross', 'Meta_score', 'Overview'], axis=1)
df_d.head()
```

```python
# 컬럼명 변경 : 
# Series_Title -> Title %

# TO-DO
df_d.rename(columns={'Series_Title':'Title'}, inplace=True)
```

```python
# Runtime 에서 min 없애고 dtype 을 int16으로 변경

# TO-DO
df_d['Runtime'] = df_d['Runtime'].str.replace('min','').str.strip().astype('int16')
df_d.info()
```

```python
df_d.head()
```

```python
# Apply method 작성
def get_stars(row):
    return f"{row.Star1}, {row.Star2}, {row.Star3}, {row.Star4}"
```

```python
# get_stars 메써드를 사용해서 apply() 적용하여 'Stars' 컬럼 생성
df_d['Stars'] = df_d.apply(get_stars, axis=1)
df_d.head()
```

```python
# 기존 컬럼 Star1, Star2, Star3, Star4 제거
df_d.drop(['Star1','Star2','Star3','Star4'], axis=1, inplace=True)
```
