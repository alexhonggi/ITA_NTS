# 0. print() 함수 사용법

print() 함수의 사용 방법
  - print(변수) : 변수 출력  
  - print(변수, 변수, 변수) : 변수 여러개 출력
  - print(문자열) : 문자열 출력
  - print(문자열, 변수, 문자열, 변수) : 문자열과 변수 섞어서 출력
  - print('문자열 {}'.format(변수)) : 따옴표 안에 {} 에 변수를 대입하여 출력
  - print(f'문자열 {변수}') : 따옴표 안에 {}에 변수를 대입하여 출력

다음을 실행하면서 print() 함수의 용법을 익혀 보자.

````python
a = 100
b = 200
c = 123.45
d = 'TEST'
````

````python
print(a)
````

````python
print(c)
````

````python
print('문자열 입니다.')
````

````python
print('a의 값은 ', a)
````

````python
print('a의 값은 ', a, ' 입니다.')
````

````python
print('a의 값은 {} 입니다.'.format(a))
````

````python
print(f'a의 값은 {a} 입니다.')
````
