※ COLAB 사용 방법
  - 공유된 COLAB 파일을 열은 다음, 파일>Drive에 사본저장 클릭하여 본인 Google Drive 로 colab 실습 파일 복제하여 실습 진행
  - 셀 : 코드 셀과 텍스트 셀로 나눠져 있음
  - 셀 삽입 : 상단의 '+ 코드', '+ 텍스트' 버튼 클릭
  - 셀 삭제 : 셀의 오른쪽 툴바의 삭제 툴바 이용
  - 셀 이동 : 셀의 오른쪽 툴바의 화살표 툴바 이용
  - 셀 실행  
     ① 실행 후 해당 셀에 위치하기 : Shift + ENTER  
     ② 실행 후 아래에 신규 셀 생성하여 신규 셀에 위치하기 : Alt + ENTER
  - 노트북 파일 저장 : Ctrl + s

# 실습해 보기

## 이 셀 하단에 "코드" 셀을 자유롭게 생성해 보세요.
- 셀을 선택할 때는 마우스로 한번 클릭 하면 됩니다.
- 실습 내용 sample
- a = 10
- b = 20
- c = a + b

## 이 셀 하단에 "텍스트" 셀을 자유롭게 생성하고 내용을 입력해 보세요.
- 텍스트 셀을 편집할 때는 더블클릭하면 됩니다.

## 위에서 생성한 "코드" 셀이나 "텍스트" 셀을 오른쪽 상단의 화살표를 이용하여 자유롭게 이동해 보세요.

## 상단의 "연결" 버튼을 클릭하여 런타임에 연결해 보세요.
- 런타임은 구글에서 제공하는 클라우드 서버를 의미합니다.
- 여러분은 따로 컴퓨터에 파이썬을 설치하지 않아도 실습을 진행 할 수 있습니다.  
  (numpy, pandas, matplotlib, scipy 등의 기본적인 라이브러리는 설치되어 있습니다.)

## "코드" 셀이나 "텍스트" 셀을 실행할 때는 SHIFT + ENTER 하거나 CTRL + ENTER 하세요.
- 차이가 무엇인지 확인해 보세요.

## 런타임 메뉴를 클릭해서 "세션관리"를 선택해 보세요.
- Colab 에 여러 세션을 실행중이면 실행중인 세션을 종료해야 한다라는 경고 메시지가 뜰 때 사용하면 됩니다.
