### 답안

```python
def alphabet_counter(input_string):
    # 초기화
    uppercase_count = 0
    lowercase_count = 0

    # 문자열 순회
    for char in input_string:
        if char.isupper():  # 대문자인 경우
            uppercase_count += 1
        elif char.islower():  # 소문자인 경우
            lowercase_count += 1

    # 결과 출력
    print("Number of uppercase letters:", uppercase_count)
    print("Number of lowercase letters:", lowercase_count)

    # 대문자 수와 소문자 수의 합 반환
    return uppercase_count + lowercase_count

### 아래 파트는 남겨둔채 위의 코드를 작성하세요.
print(alphabet_counter("Thank God It Is Friday"))
```

```python
balance = 0

def deposit():
    global balance
    how_much = int(input("How much? "))
    balance += how_much
    print("Your deposit has been approved.")

def withdrawal():
    global balance
    how_much = int(input("How much? "))
    if how_much > balance:
        print("You have insufficient funds.")
    else:
        balance -= how_much
        print("Your withdrawal has been approved.")

def atm():
    global balance
    while True:
        user_input = input("Deposit(d), withdrawal(w), or balance check(c)? ")
        if user_input == "w": withdrawal()
        elif user_input == "d": deposit()
        elif user_input == "c": 
            print(f"Your current balance is {balance}.")
        elif user_input == "":
            break
        else:
            print("Unknown command.")

atm()
```

```python
# 반복문을 사용한 피보나치 수열 계산
def fibonacci_iterative(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1

    prev, current = 0, 1
    for _ in range(2, n + 1):
        temp = current
        current = prev + current
        prev = temp

    return current

# 재귀를 사용한 피보나치 수열 계산
def fibonacci_recursive(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

### 아래의 파트는 남겨둔채 위의 코드를 작성하세요.
print(fibonacci_iterative(8))
print(fibonacci_recursive(8))
```
